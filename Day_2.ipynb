````python
# Arithmetic Operators

# +
# -
# *
# /
# % - Is sused to find the remainder
# ** - exponential , power 
# // - to get valuse before the point 4.0 -> ans 4

# a = 4
# b = 10

print(a//b)
````

````python
# Comparison Operators

# == - to check both value are equal
# != - to check the value are not equal
# > 
# <
# >=
# <=


# a = 5
# b = 10

print(a == b) #False
print(a != b) #True
print(a > b) #False
print(a < b) #True
print(a >= b) #False
print(a <= b) #True



````

````python

# Logical Operators

# and - Both are must True -> True output
# or - Any of the one True -> True output
# not

# a = 10
# b = 20

print(a <= b and a < b)
print(a <= b or a < b)
print(not True)


````

````python
# Assignment operators
# +=
# -=
# *=
# /=
# %= - Is sused to find the remainder
# **= - exponential , power 
# //= - to get valuse before the point 4.0 -> ans 4


total = 30
value = int(input("Enter the another number: "))
total += value
# total = total + value 
print(total)
````
