### Question 7

```python
num1=int(input())
num2=int(input())
op=input()
if op=='+':
    print(num1+num2)
elif op=='-':
    print(num1-num2)
elif op=='*':
    print(num1*num2)
elif op=='/':
    print(num1/num2)
else:
    print("invalid operator")
```

### Question 8
Check if a year is a leap year.

```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False
```

### Question 18
Check if a number is prime.

```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```

### Question 19
Count the number of digits in a number using loop.

```python

num = int(input("Enter a number: "))

count = 0

if num < 0:
    num = -num

if num == 0:
    count = 1
else:
    while num > 0:
        num = num // 10
        count += 1

print("Number of digits:", count)
```

### Question 20
Reverse a number using loop.

```python
num = int(input("Enter a number: "))

reverse = 0

negative = False
if num < 0:
    negative = True
    num = -num

while num > 0:
    digit = num % 10        
    reverse = reverse * 10 + digit
    num = num // 10         

if negative:
    reverse = -reverse

print("Reversed number:", reverse)
```
