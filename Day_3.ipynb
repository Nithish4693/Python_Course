#### Control Flow

```python
# IF , Elif , Else 
age = int(input("Enter your age :"))
if age > 18 : # True
    print("Your are eligible to vote !")
    
elif age > 15 :
    print("Your are at middle age person !")
    
    

else:
    print("Your are minor!!")
```
